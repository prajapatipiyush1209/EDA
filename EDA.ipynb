# Feature Selection

```python
#Shows the missing value in the data 
missing_info= pd.DataFrame(app_data.isnull().sum().sort_values().reset_index())
missing_info.rename(columns = {'index' : 'col_name', 0: "missng_values"}, inplace = True)
```

```python
#now calculate the percentage of the missing values in the data 
missing_info['missing_value_percnt'] = missing_info['missng_values']/len(app_data)*100
```

```python
missing_info
```

```python
#I want to drop the columns have missing_value_percnt is > 40 because that's not very useful in the futher analysis 
#and not imputing to missing value.
missing_col = missing_info[missing_info['missing_value_percnt'] >= 40.000]['col_name'].to_list()
app_data.drop(labels = missing_col, axis = 1, inplace= True)
```

```python
#Now after the drop the columns 
app_data.shape
```

# Feature Engineering

- Handling the Missing values

```python
app_data.info()
```

```python
#Check the missing value 
app_data.isnull().sum()
```

- First i would to handle the missing value of the categorical data.

```python
app_data['NAME_TYPE_SUITE'].value_counts(ascending= True)
```

```python
#Now here clealy shows that unaccompanied has the most frequent in the dataset then use a mode value to handle the 
#Missing value
app_data['NAME_TYPE_SUITE'] = app_data['NAME_TYPE_SUITE'].fillna(app_data['NAME_TYPE_SUITE'].mode()[0])
```

```python
app_data['OCCUPATION_TYPE'].value_counts()
```

```python
#In this case not impute the null value because if we use null value then it's not meaning of the analysis so 
#i use the mode of the column for the better analysis 

app_data['OCCUPATION_TYPE'] = app_data['OCCUPATION_TYPE'].fillna(app_data['OCCUPATION_TYPE'].mode()[0])
```

- Now handle the missing value of the numerical columns 

```python
app_data['AMT_ANNUITY'].describe()
```

```python
sns.kdeplot(app_data['AMT_ANNUITY'])
# Now here Shows that data is right skewed of the data then use the median as the fill missing values
```

```python
app_data['AMT_ANNUITY'] = app_data['AMT_ANNUITY'].fillna((app_data['AMT_ANNUITY'].median()))
```

```python
app_data['AMT_GOODS_PRICE'] = app_data['AMT_GOODS_PRICE'].fillna(app_data['AMT_GOODS_PRICE'].median())
```

```python
app_data['CNT_FAM_MEMBERS'].value_counts()
```

```python
app_data['CNT_FAM_MEMBERS'] = app_data['CNT_FAM_MEMBERS'].fillna('2.0')
```

```python
for col in app_data.columns:
    if col.startswith('AMT_REQ_CREDIT_BUREAU_'):
        app_data[col] = app_data[col].fillna(app_data[col].median())
```

```python
app_data['OBS_30_CNT_SOCIAL_CIRCLE'] = app_data['OBS_30_CNT_SOCIAL_CIRCLE'].fillna(app_data['OBS_30_CNT_SOCIAL_CIRCLE'].median())
app_data['DEF_30_CNT_SOCIAL_CIRCLE'] = app_data['DEF_30_CNT_SOCIAL_CIRCLE'].fillna(app_data['DEF_30_CNT_SOCIAL_CIRCLE'].median())
app_data['OBS_60_CNT_SOCIAL_CIRCLE'] = app_data['OBS_60_CNT_SOCIAL_CIRCLE'].fillna(app_data['OBS_60_CNT_SOCIAL_CIRCLE'].median())
app_data['DEF_60_CNT_SOCIAL_CIRCLE'] = app_data['DEF_60_CNT_SOCIAL_CIRCLE'].fillna(app_data['DEF_60_CNT_SOCIAL_CIRCLE'].median())
```

```python
#Remove EXT_SOURCE_3 AND EXT_SOURCE_2 from the dataset
app_data.drop(labels = ['EXT_SOURCE_2','EXT_SOURCE_3'], axis = 1, inplace = True)
```

```python
# Remove the one row from the DAYS_LAST_PHONE_CHANGE missing value
app_data.dropna(subset = ['DAYS_LAST_PHONE_CHANGE'], axis = 0, inplace = True)
```

# Univarieate Analysis

```python
plt.figure(figsize=(25,80))
for i, col_name in enumerate(object_col):

    data_pct = app_data[[col_name,'TARGET']].groupby([col_name], as_index=False).mean().sort_values(by='TARGET',ascending=False)
    data_pct['Target_Percentage'] = data_pct['TARGET']*100


    plt.subplot(11,2,i+i+1)
    plt.subplots_adjust(wspace=0.1,hspace=1)
    ax = sns.countplot(data=app_data,x=col_name,hue='TARGET')
    for p in ax.patches:
        ax.annotate(format(p.get_height(), '.0f'), 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha = 'center', va = 'center', 
                    xytext = (0, 10), 
                    textcoords = 'offset points')
   
    plt.xticks(rotation=90)

    plt.subplot(11,2,i+i+2)
    ax = sns.barplot(data=data_pct,x=col_name,y='Target_Percentage',palette='coolwarm')
    for p in ax.patches:
        ax.annotate(format(p.get_height(), '.0f'), 
                    (p.get_x() + p.get_width() / 2., p.get_height()), 
                    ha = 'center', va = 'center', 
                    xytext = (0, 10), 
                    textcoords = 'offset points')
   
    plt.xticks(rotation=90)

plt.tight_layout()
plt.show()
```
